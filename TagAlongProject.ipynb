```python
#14

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['stabf']= label_encoder.fit_transform(df['stabf'])
df['stab']= label_encoder.fit_transform(df['stab'])

X = df.drop(["stab","stabf"], axis = 1)
y = df[["stab","stabf"]]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.3 , random_state= 1 )

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 5, max_depth = 4, max_features = 3, bootstrap = True, random_state = 18).fit(x_train, y_train)
prediction = clf.predict(x_test)

from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
new_predictions = clf.predict(x_test)
cnf_mat = confusion_matrix(y_true=y_test, y_pred=new_predictions, labels=[ '2A' , '3A' ])
cnf_mat

from sklearn.metrics import accuracy_score
accuracy_score(y_test, new_predictions)
```
